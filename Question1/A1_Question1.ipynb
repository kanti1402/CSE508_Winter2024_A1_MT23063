**LIBRARY / PACKAGE**

```python
import os
import string
import re
import nltk

from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')
```

**INPUT FOLDER**

```python
input_folder = "/content/drive/MyDrive/text_files"
os.chdir(input_folder)
```

**FUNCTION FOR PRINT 5 SAMPLE FILE**

```python
def Print_File(folder_path):
    file_list = [f"file{i}.txt" for i in range(1, 1000)]
    for i in range(5):
        file_path = os.path.join(folder_path, file_list[i])
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"Contents of {file_list[i]}:\n{content}\n")
```

**FUNCTION FOR REMOVE HTML TAG**

```python
def remove_html_tags(text):
    try:
        soup = BeautifulSoup(text, "html.parser")
        return soup.get_text()
    except Exception as e:
        print(f"Error removing HTML tags: {e}")

def process_files():
    try:
        current_folder = os.getcwd()

        for file_name in os.listdir(current_folder):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                cleaned_content = remove_html_tags(content)
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(cleaned_content)
            print(f"Processed: {file_name}")

        print("HTML Tag Remopve Successfully.")
    except Exception as e:
        print(f"Error listing files in directory: {e}")

process_files()
```

**FUNCTION FOR LOWERCASE**

```python
def convert_to_lowercase():
    try:
        current_folder = os.getcwd()

        for file_name in os.listdir(current_folder):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r') as file:
                content = file.read().lower()

            with open(file_path, 'w') as file:
                file.write(content)

        print("Conversion to lowercase completed successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

**FUNCTION FOR TOKENIZATION**

```python
def tokenize():
    try:
        current_folder = os.getcwd()

        for i, file_name in enumerate(os.listdir(current_folder), start=1):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r') as file:
                data = nltk.word_tokenize(file.read())

                if i <= 5:
                    print(f"\nFile {i}\n")
                    print(data)

            with open(file_path, 'w') as file:
                file.write(' '.join(data))

        print("Tokenization completed successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

**FUNCTION FOR REMOVE STOP WORD**

```python
def remove_stop_word():
    try:
        current_folder = os.getcwd()

        for file_name in os.listdir(current_folder):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r') as file:
                stop_words = set(stopwords.words('english'))
                tokens = nltk.word_tokenize(file.read())
                filtered_words = [w for w in tokens if w.lower() not in stop_words]
                new_sentence = ' '.join(filtered_words)

            with open(file_path, 'w') as file:
                file.write(new_sentence)

        print("Stop words removal completed successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

**FUNCTION FOR REMOVE PUNCTUATIONS**

```python
def remove_punctuations():
    try:
        current_folder = os.getcwd()

        for file_name in os.listdir(current_folder):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r') as file:
                content = file.read()
                clean_content = content.translate(str.maketrans('', '', string.punctuation))

            with open(file_path, 'w') as file:
                file.write(clean_content)

        print("Punctuation removal completed successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

**FUNCTION FOR REMOVE BLANK SPACE**

```python
def remove_blank_space():
    try:
        current_folder = os.getcwd()

        for file_name in os.listdir(current_folder):
            file_path = os.path.join(current_folder, file_name)

            with open(file_path, 'r') as file:
                content = file.read()
                cleaned_content = re.sub(r"\s+", " ", content, flags=re.UNICODE)

            with open(file_path, 'w') as file:
                file.write(cleaned_content)

        print("Blank space removal completed successfully.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

**Q1. A: Lowercase the text**

```python
if __name__ == "__main__":
    print("Before Lowercase:\n")
    Print_File(input_folder)
    print("===============================================================================================================")
    convert_to_lowercase()
    print("LowerCase conversion completed.")
    print("===============================================================================================================")
    print("After Lowercase:\n")
    Print_File(input_folder)
    print("===============================================================================================================")
```
